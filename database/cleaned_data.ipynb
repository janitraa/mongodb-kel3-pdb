# **Pra Pemrosesan Data Kategorikal**

```
# Impor library untuk membantu pra pemrosesan data kategorik ke numerik
from sklearn import preprocessing 

# Mendefinisikan fungsi LabelEncoder()
le = preprocessing.LabelEncoder()

# Mendefinisikan data pada kolom apa saja yang akan di transform ke numerik
for column in ['ID', 'Zipcode', 'Sunrise_Sunset', 'Civil_Twilight', 'Nautical_Twilight', 'Astronomical_Twilight;;;', 'Traffic_Signal']:
    le.fit(df[column])
    df[column] = le.transform(df[column])
```

```
# Memeriksa data yang sudah diubah ke dalam bentuk numerik
df.dtypes
```

# **Data Refinement**

```
# Mendefinisikan kolom dengan nilai modus > 90%
mode_cols = []
for i in df.columns:
  if len(df[df[i] == df[i].mode().get(0)])/df.shape[0]>0.9:
    mode_cols.append(i)
mode_cols
```

```
# Drop kolom yang memenuhi kondisi diatas
df.drop(mode_cols, axis=1, inplace=True)
```

```
# Periksa dataset setelah kolom dihapus
df.dtypes
```

```
# Mengubah format kolom Start_Time dan End_Time dalam bentuk datetime
df.Start_Time = pd.to_datetime(df.Start_Time)
df.End_Time = pd.to_datetime(df.End_Time)
```

```
# Membagi kolom-kolom berdasarkan tipe waktu (Tanggal, jam, bulan, tahun)
df['date_start'] = df['Start_Time'].dt.date
df['time_start'] = df['Start_Time'].dt.time
df['month_start'] = df['Start_Time'].dt.month
df['year_start'] = df['Start_Time'].dt.year
df['date_end'] = df['End_Time'].dt.date
df['time_end'] = df['End_Time'].dt.time
df['month_end'] = df['End_Time'].dt.month
df['year_end'] = df['End_Time'].dt.year
```

```
# Memeriksa data baru yang telah ditamnbahkan
df.head()
```

```
# Mendefinisikan nilai dari kolom atribut Total
df['Total']= df['Sunrise_Sunset'] + df['Civil_Twilight'] + df['Nautical_Twilight'] + df['Astronomical_Twilight;;;']
```

```
# Mendefinisikan fungsi period, yang menentukan atribut Period_Time
def period(total): 
  df.loc[(total/4 >= 0.5),['Period_Time']] = 1
  df.loc[(total/4 < 0.5),['Period_Time']] = 0

# Memasukan nilai kolom atribut Period_Time
period(df['Total'])
```

```
# Menghapus outlier
for column in ['Temperature(F)','Humidity(%)','Pressure(in)','Visibility(mi)']:
  Q1 = df[column].quantile(0.25)
  Q3 = df[column].quantile(0.75)
  IQR = Q3 - Q1    

  filter = (df[column] >= Q1 - 1.5 * IQR) & (df[column] <= Q3 + 1.5 *IQR)
  df.loc[filter] 
```

# **Mengolah JSON**

```
# Mendefinisikan json dataset
df_json = df.to_json(orient='records')
```

```
# Mengimpor library json dan memasukan dataset yang sudah diubah 
import json
df_load_json = json.loads(df_json)
```

```
# Membuat dataset baru untuk data berbentuk json
df_new =[]
count= 0
```

```
df.head()
```

```
# Memasukan data json ke dataset baru
for i in df_load_json:
    name={
        '_id': i['_id'],
        'id_accident': i['id_accident'],
        'severity':i['severity'],
        'location':{
            'coordinates':{
                'start_lng': i['start_lng'],
                'start_lat': i['start_lat']
                },
            'type':"Point"
        },
        'side': i['side'],
        'city': i['city'],
        'county': i['county'],
        'state':i['state'],
        'zipcode': i['zipcode'],
        'temperature': i['temperature'],
        'humidity': i['humidity'],
        'pressure':i['pressure'],        
        'visibility': i['visibility'],
        'wind_speed': i['wind_speed'],
        'weather_condition': i['weather_condition'],
        'traffic_signal': i['traffic_signal'],
        'date_start': i['date_start'],
        'time_start': i['time_start'],
        'month_start':i['month_start'],  
        'year_start':i['year_start'],            
        'time_end': i['time_end'],
        'period_time':i['period_time']  
    }
    df_new.insert(count,name)
    count+=1
```

```
# Melihat isi dari dataset baru
df_new
```

```
len(df_new)
```

```
# Menulis file json
with open('final.json', 'w') as json_file:
    for row in df_new:
        json_file.write(json.dumps(row,indent=4))
```
